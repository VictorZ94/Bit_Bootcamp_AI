```python
import skfuzzy as fuzz
import numpy as np
import matplotlib.pyplot as plt
```

### Variables: 

- Tendencia precio reciente -> Universo del discurso: (-10%,10%) %: Sube fuerte, Estable, Baja Fuerte.
- Volumen de negociaciÃ³n -> Universo del discurso: (0,1000) miles: Bajo, Media, Alta.
- Sentimiento del mercado -> Universo del discurso: (0,100) miles "Negativo", "Neutral", "Positivo".

```python
x_tendencia = np.arange(-10, 10.1, 0.1)
x_volumen = np.arange(0, 1000, 1)
x_RSI = np.arange(0, 100, 1)
```

```python
# Tendencia BAJA (Precio cayendo fuerte)
tendencia_baja = fuzz.trimf(x_tendencia, [-10, -5, 0])

# Tendencia NEUTRAL (Estable o cambios ligeros)
tendencia_media = fuzz.trimf(x_tendencia, [-2, 0, 2])

# Tendencia ALTA (Precio subiendo fuerte)
tendencia_alta = fuzz.trimf(x_tendencia, [0, 5, 10])
```

```python
plt.plot(x_tendencia, tendencia_baja, 'y', linewidth=1.5, label='Baja')
plt.plot(x_tendencia, tendencia_media, 'black', linewidth=1.5, label='Media')
plt.plot(x_tendencia, tendencia_alta, 'v', linewidth=1.5, label='Larga')
```

```python
volumen_bajo = fuzz.trimf(x_volumen, [0, 250, 500])
volumen_estable = fuzz.trimf(x_volumen, [400, 500, 600])
volumen_alto = fuzz.trimf(x_volumen, [550, 750, 1000])

plt.plot(x_volumen, volumen_bajo, 'y', linewidth=1.5, label='Bajo')
plt.plot(x_volumen, volumen_estable, 'black', linewidth=1.5, label='Estable')
plt.plot(x_volumen, volumen_alto, 'v', linewidth=1.5, label='Alto')
```

```python
RSI_sobrevendida = fuzz.trimf(x_RSI, [0, 2500, 5000])
RSI_neutral = fuzz.trimf(x_RSI, [4000, 5000, 6000])
RSI_sobrecomprada = fuzz.trimf(x_RSI, [5500, 7500, 10000])

plt.plot(x_RSI, RSI_sobrevendida, 'y', linewidth=1.5, label='Sobrevendida')
plt.plot(x_RSI, RSI_neutral, 'black', linewidth=1.5, label='Neutral')
plt.plot(x_RSI, RSI_sobrecomprada, 'v', linewidth=1.5, label='Sobrecomprada')

plt.title('Funciones de pertenencia')
plt.legend()
plt.show()
```

# Segunda parte

```python
# Universo de discursos 
x = np.linspace(0,10,1000)

A = fuzz.gaussmf(x,3,1)
B = fuzz.gaussmf(x,6,1)
C = fuzz.sigmoid(x,5,8)
```

```python
plt.plot(x,A,'r',linewidth=2,label='A')
plt.plot(x,B,'g',linewidth=2,label='B') 
plt.plot(x,C,'b',linewidth=2,label='C')
plt.title('Funciones de membresÃ­a')
plt.show()
```

```python
#Operaciones de conjuntos
#Complemento 
Ac = 1 - A
Bc = 1 - B
Cc = 1 - C

plt.plot(x,Ac,'m',linewidth=2,label='Ac')
plt.plot(x,Bc,'y',linewidth=2,label='Bc')
plt.plot(x,Cc,'c',linewidth=2,label='Cc')
plt.title('Complemento de A')
plt.show()
```

```python
plt.plot(x,Bc,'m',linewidth=2,label='Bc')
plt.title('Complemento de B')
```

```python
plt.plot(x,Cc,'m',linewidth=2,label='Cc')
plt.title('Complemento de C')
```

```python
BcuCc = np.fmax(Bc,Cc)
```

```python
plt.plot(x,BcuCc,'black', linewidth=2,label='AuB')
plt.title('Union de B complement y C complemento')
plt.legend()
plt.show()
```

```python
BcuCcnA = np.fmin(BcuCc,A)
```

```python
plt.plot(x,BcuCcnA,'g',linewidth=2,label='BcuCcnA')
plt.title('B complemento unido con C complement inter A')
plt.legend()
plt.show()
```
