```python
# Loggica Difusa

#pip install sckit-fuzzy
import skfuzzy as fuzz 
import numpy as np
import matplotlib.pyplot as plt
```

### Variables: 

- Velocidad -> Universo del discurso: (0,101) km/h: Velocidad Alta, Velocidad Media, Velocidad Baja
- Distancia -> Universo del discurso: (0,121) km: Distancia corta, distancia media, distancia lejana

```python
x_distancia = np.arange(0, 101, 1)
x_velocidad = np.arange(0, 121, 1)
```

```python
distancia_corta = fuzz.trimf(x_distancia, [0,25,40])
distancia_media = fuzz.trimf(x_distancia, [20,35,50])
distancia_larga = fuzz.trimf(x_distancia, [37,70,100])

plt.plot(x_distancia, distancia_corta, 'b', linewidth=1.5, label='Corta')
plt.plot(x_distancia, distancia_media, 'g', linewidth=1.5, label='Media')
plt.plot(x_distancia, distancia_larga, 'r', linewidth=1.5, label='Larga')
plt.show()
```

```python
velocidad_lenta = fuzz.trimf(x_velocidad, [0,30,60])
velocidad_media = fuzz.trimf(x_velocidad, [30,50,70])
velocidad_alta = fuzz.trimf(x_velocidad, [50,90,120])

plt.plot(x_velocidad, velocidad_lenta, 'b', linewidth=1.5, label='Lenta')
plt.plot(x_velocidad, velocidad_media, 'g', linewidth=1.5, label='Media')
plt.plot(x_velocidad, velocidad_alta, 'r', linewidth=1.5, label='Alta')
```
