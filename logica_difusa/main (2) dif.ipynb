```python
import numpy  as np 
import skfuzzy as fuzz 
import matplotlib.pyplot as plt
```

```python
#Universo de discursos 
x = np.linspace(0,10,1000)

A = fuzz.trapmf(x,[0,1,4,5])
B = fuzz.trimf(x,[2,5,7])
C = fuzz.gaussmf(x,7,1)

plt.plot(x,A,'r',linewidth=2,label='A')
plt.plot(x,B,'g',linewidth=2,label='B') 
plt.plot(x,C,'b',linewidth=2,label='C')
plt.title('Funciones de membres√≠a')
plt.show()
```

```python
#Operaciones de conjuntos
#Complemento 
Ac = 1 - A
Bc = 1 - B
Cc = 1- C

plt.plot(x,Ac,'m',linewidth=2,label='Ac')
plt.plot(x,Bc,'y',linewidth=2,label='Bc')
plt.plot(x,Cc,'c',linewidth=2,label='Cc')
plt.title('Complemento de A')
plt.show()
```

```python
#Union  max 
AuB = np.fmax(A,B)
BuC = np.fmax(B,C)
AuC = np.fmax(A,C)
```

```python
plt.plot(x,AuB,'m',linewidth=2,label='AuB')
plt.title('Union de A y B')
plt.legend()
plt.show()
```

```python
plt.plot(x,BuC,'y',linewidth=2,label='BuC')
plt.title('Union de B y C')
plt.legend()
plt.show()
```

```python
plt.plot(x,AuC,'c',linewidth=2,label='AuC')
plt.title('Union de A y C')
plt.legend()
plt.show()
```

```python
AuBuC = np.fmax(AuB,C)
plt.plot(x,AuBuC,'r',linewidth=2,label='AuBuC')
plt.title('Union de A,B y C')
plt.legend()
plt.show()
```

```python
AuBiC = np.fmin(AuB,C)
plt.plot(x,AuBiC,'r',linewidth=2,label='AuBiC')
plt.title('Intersecci√≥n de A y C')
plt.legend()
plt.show()
```
